```python
import pandas as pd
```

```python
filename = 'Infrastructure investment'
data_frame = pd.read_csv(filename+'.csv')
```

```python
data_frame
```

```python
data_frame['SUBJECT'].unique()
```

```python
data_frame['INDICATOR'].unique()
```

```python
data_frame['LOCATION'].unique()
```

```python
data_frame['TIME'].unique()
```

```python
investment_data = data_frame[["LOCATION","SUBJECT","TIME","Value"]]
```

```python
investment_data
```

```python
for subject in investment_data['SUBJECT'].unique():
    subject_data = investment_data[investment_data['SUBJECT'] == subject]
    subject_data = subject_data.pivot_table(columns='TIME',index='LOCATION',values='Value',aggfunc='mean')
    subject_data.to_csv(filename+subject+".csv")
```
